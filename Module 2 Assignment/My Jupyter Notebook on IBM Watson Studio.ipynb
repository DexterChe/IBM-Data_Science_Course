### The code below plots the sin(x) and cos(x) function.
1. Import of libraries
2. Make equally spaced values of x in the range from 1 to 10
3. Calculate sin(x) and cos(x) values for each of x
4. Plot functions on the same plot

````python
import numpy as np
import matplotlib.pylab as plt

x = np.linspace(0, 10, num=10000)
y1 = np.sin(x)
y2 = np.cos(x)

fig,axs = plt.subplots(1,1, figsize=(3,2), dpi=150)
axs.plot(x,y1, linewidth=1, label='sin(x)')
axs.plot(x, y2, linewidth=1, label='cos(x)')
axs.set_title('Trigonometric functions')
axs.set_xlabel('x', weight='bold')
axs.set_ylabel('y', weight='bold')
axs.legend(loc='upper right')
plt.show()
````
